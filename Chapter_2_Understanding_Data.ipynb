# Z score

```python
# prompt: do min-max normalization for the df2 for each column separately

def min_max_normalize(df, column_name):
    """
    Performs min-max normalization on a specified column of a DataFrame.

    Args:
        df: The input DataFrame.
        column_name: The name of the column to normalize.

    Returns:
        A pandas Series containing the min-max normalized values for the specified column.
        Returns None if the column is not found in the DataFrame or if there's an error during normalization.
    """
    if column_name not in df.columns:
        print(f"Error: Column '{column_name}' not found in DataFrame.")
        return None

    min_val = df[column_name].min()
    max_val = df[column_name].max()

    if min_val == max_val:  # Handle cases where all values are the same
        print(f"Warning: All values in column '{column_name}' are the same. Returning 0 for all normalized values.")
        return pd.Series(0, index=df.index)

    normalized_column = (df[column_name] - min_val) / (max_val - min_val)
    return normalized_column

# Example usage (replace with your actual DataFrame and column names)
# Assuming df2 is defined in your code as in the provided file

# Normalize 'X1' column of df2
normalized_X1 = min_max_normalize(df2, 'X1')
print(f"Min-Max Normalized 'X1':\n{normalized_X1}\n")

# Normalize 'X2' column of df2
normalized_X2 = min_max_normalize(df2, 'X2')
print(f"Min-Max Normalized 'X2':\n{normalized_X2}\n")


#To normalize all columns in df2:
for column in df2.columns:
    normalized_column = min_max_normalize(df2, column)
    if normalized_column is not None:
        print(f"Min-Max Normalized '{column}':\n{normalized_column}\n")
```
