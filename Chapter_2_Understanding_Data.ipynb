# Distance

```python
# prompt: build pd data frame from  [[1,2],[-1,4],[0,3],[-3,2]] column 1 name X1 column 2 name  X2

import pandas as pd

data = [[1, 2], [-2, 4], [0, 3], [-3, -2]]
df = pd.DataFrame(data, columns=['X1', 'X2'])
df
```

```python
# prompt: draw scatter plot of X1 X2 with line 0 in x and y

import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame with 'X1' and 'X2' columns
# Replace this with your actual DataFrame if it's named differently

# Create the scatter plot
plt.figure(figsize=(8, 6))  # Adjust figure size if needed
plt.scatter(df['X1'], df['X2'])

# Add a line at x=0 and y=0
plt.axhline(0, color='red', linestyle='--', linewidth=1)
plt.axvline(0, color='red', linestyle='--', linewidth=1)


# Add labels and title
plt.xlabel("X1")
plt.ylabel("X2")
plt.title("Scatter Plot of X1 vs X2")

# Display the plot
plt.grid(True)
plt.show()
```

```python
# prompt: calculate distance matrix of the df

import pandas as pd
from scipy.spatial.distance import pdist, squareform

# Assuming 'df' is your DataFrame with relevant columns for distance calculation
# Replace 'X1' and 'X2' with the actual column names if different

# Calculate the pairwise distances between rows
distances = pdist(df[['X1', 'X2']], metric='euclidean')  # You can change the metric if needed

# Convert the condensed distance matrix to a square distance matrix
distance_matrix = squareform(distances)

# Create a DataFrame from the distance matrix
distance_df = pd.DataFrame(distance_matrix, index=df.index, columns=df.index)

distance_df
```

```python
# prompt: calculate distance matrix of the df use L1-norm

# Assuming 'df' is your DataFrame with relevant columns for distance calculation
# Replace 'X1' and 'X2' with the actual column names if different

from scipy.spatial.distance import pdist, squareform

# Calculate the pairwise distances between rows using L1 norm (Manhattan distance)
distances = pdist(df[['X1', 'X2']], metric='cityblock')

# Convert the condensed distance matrix to a square distance matrix
distance_matrix = squareform(distances)

# Create a DataFrame from the distance matrix
distance_df = pd.DataFrame(distance_matrix, index=df.index, columns=df.index)

distance_df
```

```python

# prompt: calculate distance matrix of the df2

# Assuming df2 is your DataFrame with 'X1' and 'X2' columns
# Replace this with your actual DataFrame if it's named differently

from scipy.spatial.distance import pdist, squareform
import pandas as pd

# Calculate the pairwise distances between rows using Euclidean distance
distances = pdist(df2[['X1', 'X2']], metric='euclidean')

# Convert the condensed distance matrix to a square distance matrix
distance_matrix = squareform(distances)

# Create a DataFrame from the distance matrix
distance_df = pd.DataFrame(distance_matrix, index=df2.index, columns=df2.index)

distance_df
```

# Normalization (Z-score & Min-Max normalization)

```python

# prompt: calculate z-score of df2

# Assuming df2 is your DataFrame with 'X1' and 'X2' columns
# Replace this with your actual DataFrame if it's named differently

from scipy.stats import zscore

# Calculate z-scores for 'X1' and 'X2' columns
df2['X1_zscore'] = zscore(df2['X1'])
df2['X2_zscore'] = zscore(df2['X2'])

df2
```

```python

# prompt: draw scatter plot df2 of X1 X2 with line 0 in x and y

# Assuming df2 is your DataFrame with 'X1' and 'X2' columns
# Replace this with your actual DataFrame if it's named differently

# Create the scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(df2['X1_zscore'], df2['X2_zscore'])

# Add a line at x=0 and y=0
plt.axhline(0, color='red', linestyle='--', linewidth=1)
plt.axvline(0, color='red', linestyle='--', linewidth=1)

# Add labels and title
plt.xlabel("X1")
plt.ylabel("X2")
plt.title("Scatter Plot of X1 vs X2")

# Display the plot
plt.grid(True)
plt.show()
```

```python

# prompt: do min-max normalization for the df2 for each column separately

# Assuming df2 is your DataFrame with 'X1' and 'X2' columns
# Replace this with your actual DataFrame if it's named differently

# Apply min-max normalization to each column separately
for column in df2.columns:
    min_val = df2[column].min()
    max_val = df2[column].max()
    df2[column + '_minmax'] = (df2[column] - min_val) / (max_val - min_val)

df2
```

```python
# prompt: draw scatter plot df2 of X1 X2 with line 0 in x and y

# Assuming df2 is your DataFrame with 'X1' and 'X2' columns
# Replace this with your actual DataFrame if it's named differently

# Create the scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(df2['X1_minmax'], df2['X2_minmax'])

# Add a line at x=0 and y=0
plt.axhline(0, color='red', linestyle='--', linewidth=1)
plt.axvline(0, color='red', linestyle='--', linewidth=1)

# Add labels and title
plt.xlabel("X1")
plt.ylabel("X2")
plt.title("Scatter Plot of X1 vs X2")

# Display the plot
plt.grid(True)
plt.show()

## Data point จาก Min-Max Normalization ส่วนใหญ่จะอยู่ใน Quadrant ที่ 1
## การแบ่งกลุ่ม จะแบ่งกลุ่ม Data ด้วยระยะห่างของ Data
```
